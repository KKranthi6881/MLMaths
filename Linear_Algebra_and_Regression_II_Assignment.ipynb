 Write a Python function that takes a matrix A and an integer k as input and returns the top k principal components of A using PCA.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets as datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
def k_Complonents_of_PCA (A,k):
    #Standartizing the Matrix values 0 - 1 with mean avg

    scaler = StandardScaler()
    Xst = scaler.fit_transform(A)

    # Defining the PCA values
    pca = PCA(n_components = k)
    xDf_PCA = pca.fit(Xst)
    PCA_Tranformed = pca.fit_transform(Xst)
    max_variance = max(xDf_PCA.explained_variance_ratio_)
    variance_value_orders = xDf_PCA.explained_variance_ratio_
    std = print(f"stanrdatized values : ",Xst)
    #return the Maximum variance ratio value
    return variance_value_orders,max_variance,PCA_Tranformed
```

```python
#  Defining the Matrix value
A=np.array([[2,3,4],[2,3,4],[4,5,6]])
k = 2
variance_value_orders,max_variance,PCA_Tranformed  = k_Complonents_of_PCA (A,k)
print(f"List of Transformed PCA Components",PCA_Tranformed)
print(f" List of {k} PCA Components values",variance_value_orders)
print(f" Top PCA variance value",max_variance)
```

**Plotting the vatiance ratio with PCA**




```python
plt.scatter(x = [i + 1 for i in range(len(variance_value_orders))],
            y = variance_value_orders)
plt.grid(True)
plt.title("Explained variance ratio of the fitted principal component vector")
plt.xlabel("Principal components")
plt.xticks([i+1 for i in range(len(variance_value_orders))])
plt.yticks()
plt.ylabel("Variance ratio")
plt.show()
```
