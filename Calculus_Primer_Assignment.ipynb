## **Q1) Implement the following function:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the cost function for the enerrgy consumption system
def J(x):
  return x ** 2 - 4 * x +3

# Define the Gradient of J(x)
def grad_J(x):
  return 2 * x - 4


#Define the Gradient Descent function

def gradient_descent(J, grad_J, x, alpha, num_iters):



    """

    Performs Gradient Descent to optimize the cost function J(theta)

    Arguments:
    x -- initial parameter value (scalar)
    alpha -- learning rate (scalar)
    num_iters -- the number of iterations to run Gradient Descent (scalar)

    Returns:
    x_final -- optimized parameter value
    J_history -- vector of cost function values for each iteration (num_iters x 1)

    """

    # Define the Cost Function
    J_history = [J(x)]

    # Define the temperature parameter value
    x_list = [x]
    for i in range(num_iters):
      J_value = J(x)
      J_history.append(J_value)
      x = x - alpha * grad_J(x)
      x_list.append(x)


    print("Final Temperature value",x)



    return x_list, J_history

# Learning rate
alpha = 0.1

# Number of iterations
num_iters = 10
# Initializing the Temperature value
x = -5

# Gradient Descent function
x_list,J_history = gradient_descent(J, grad_J, x, alpha, num_iters)



# Creating a range of x values (temperature values)
X_range = np.linspace(-5,25,100)
y_range = J(X_range)

# Plot the cost function
plt.figure(figsize=(10, 5))
plt.plot(X_range, y_range, label='Cost Function J(x)')
plt.scatter(x_list, J_history, color='red', label='Gradient Descent Path')
plt.xlabel('x')
plt.ylabel('J(x)')
plt.title('Gradient Descent for J(x) = x^2 - 4x + 3')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()

```
